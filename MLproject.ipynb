```python
#Loading the dataset

# here we add sep='\t' as in csv(comma-seperated-values) defaut is comma but r data is tab seperated so mention \t
df = pd.read_csv("marketing_campaign (2).csv", sep = '\t')
print(df.shape)
```

```python
# removing null values (since 2240 total rows but 'income' column has 2216 rows so this column has null values)

df = df.dropna()
print(df.shape)
```

```python
df['Dt_Customer'] = pd.to_datetime(df['Dt_Customer'], dayfirst = True)

dates = []
for i in df['Dt_Customer']:
    i = i.date()
    dates.append(i)

#printing dates of newest and oldest recorded customer
print("The newest customer's enrolment date in records:", max(dates))
print("The oldest customer's enrolment date in records:", min(dates))
```
