```python
import pandas as pd
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.model_selection import train_test_split
```

```python
data = {
    'StudyHours' : [1,2,3,4,5,6],
    'TestScore' : [40,50,60,70,80,90]
}

df = pd.DataFrame(data)
```

```python
std_scaler = StandardScaler()
std_scaled = std_scaler.fit_transform(df)
```

```python
print("Standar Scaler Output")
print(pd.DataFrame(std_scaled, columns = ['StudyHours' ,'TestScore']))
```

```python
print(std_scaled)
```

```python
minmax_scaler = MinMaxScaler()
minmax_scaled = minmax_scaler.fit_transform(df)

print("\n Min Max Scaler Output")
print(pd.DataFrame(minmax_scaled, columns = ['StudyHours', 'TestScore']))
```

```python
X = df[["StudyHours"]]
Y = df[["TestScore"]]
X_train, X_test, y_train, y_test = train_test_split(X,Y, test_size=0.2, random_state=42)
```
